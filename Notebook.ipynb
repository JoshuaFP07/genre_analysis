<h2>Best Released Year</h2>
<h3>1. Top 5 Released Year</h3>
<p>Since the focus of this project is to see the most successful movies and pinpoint the best genres from those movies, first we want to see some of the top years off all time based on IMDb rating and Meta Score. </p>

```python
%%sql

SELECT
  released_year,
  round(AVG(imdb_rating), 2) AS avg_imdb_rating,
  COUNT(*) AS count_movie 
FROM
  imdb_top_1000 
GROUP BY
  released_year 
ORDER BY
  avg_imdb_rating DESC LIMIT 5;
```

```python
%%sql

SELECT
  released_year,
  round(AVG(meta_score), 2) AS avg_meta_score,
  COUNT(*) AS count_movie 
FROM
  imdb_top_1000
GROUP BY
  released_year 
ORDER BY
  avg_meta_score DESC LIMIT 5;
```

<h3>2. Top 5 Released Year with More Movies based on Market Trend</h3>
<p>When we look at the years movies were released, we see that some years only have a few movies, like 1 to 3. This can mess up our calculations because there aren't enough movies to get a good insight. So, we need to focus on years where there are more movies that have been reviewed to make our analysis better. Because of that, I decided to include years that have more than 8 movies to make sure we have a more reliable analysis, based on the average of IMDb rating and Meta Score.</p>
<p>Not only that, to enhance its alignment with current trends and audiences, I put a condition where the film must be released at least in 2010.</p>

```python
%%sql

SELECT
  released_year,
  round(AVG(imdb_rating), 2) AS avg_imdb_rating,
  COUNT(*) AS count_movie 
FROM
  imdb_top_1000
WHERE 
  released_year >= 2010 
GROUP BY
  released_year 
HAVING
  COUNT(*) > 8 
ORDER BY
  avg_imdb_rating DESC LIMIT 5
```

```python
%%sql

SELECT
  released_year,
  round(AVG(meta_score), 2) AS avg_meta_score,
  COUNT(*) AS count_movie 
FROM
  imdb_top_1000
WHERE 
  released_year >= 2010
GROUP BY
  released_year 
HAVING
  COUNT(*) > 8 
ORDER BY
  avg_meta_score DESC LIMIT 5
```

<h3>3. Joining the Best Years</h3>
<p>Now if we join both years that we got from best years of IMDb Rating and best years from Meta Score, we will got a list of best released years from both list. The list consists of 3 released years. Both queries is link using CTEs to get a deep understanding about the best movie year from both user and reviewer perspective. By doing this, i believe that im not only assesing popularity but the quality of the movies. </p>

```python
%%sql

WITH trend_imdb_rating AS 
(-- Best movies from user rating
  SELECT
    released_year,
    round(AVG(imdb_rating), 2) AS avg_imdb_rating,
    COUNT(*) AS count_movie 
  FROM
    imdb_top_1000
  WHERE 
    released_year >= 2010 
  GROUP BY released_year 
  HAVING COUNT(*) > 8 
  ORDER BY avg_imdb_rating DESC LIMIT 5
)
,
trend_meta_score AS 
(-- Best movies from critics score
  SELECT
    released_year,
    round(AVG(meta_score), 2) AS avg_meta_score,
    COUNT(*) AS count_movie 
  FROM
    imdb_top_1000
  WHERE 
    released_year >= 2010 
  GROUP BY released_year 
  HAVING COUNT(*) > 8 
  ORDER BY avg_meta_score DESC LIMIT 5
)
SELECT
  ti.released_year 
FROM
  trend_imdb_rating AS ti 
  INNER JOIN
    trend_meta_score AS tm 
    ON ti.released_year = tm.released_year 
ORDER BY avg_imdb_rating DESC, avg_meta_score DESC
```

<h3>4. Best Genre</h3>
<p>Before creating the table, we need to take each movie's multiple genres and turn them into a single genre. We can do this using the <code>SUBSTRING_INDEX</code> function, which helps us join them into a list.</p>

```python
%%sql

SELECT
  movie_title,
  SUBSTRING_INDEX(SUBSTRING_INDEX(genre, ', ', n.digit + 1), ', ', - 1) AS genre -- Extract each genre from the comma-separated list
FROM
  imdb_top_1000 
  JOIN
    (
      SELECT 0 AS digit 
      UNION ALL
      SELECT 1 
      UNION ALL
      SELECT 2 
      UNION ALL
      SELECT 3
    )
    AS n -- Generates a sequence of numbers the maximum number of genres
    ON LENGTH(genre) - LENGTH(REPLACE(genre, ',', '')) >= n.digit LIMIT 10;
```

<p>With that query, now we can get the amount of genre and the total revenue for each genre </p>

```python
%%sql

WITH trend_imdb_rating AS 
(
  -- Best movies from user rating
  SELECT
    released_year,
    round(AVG(imdb_rating), 2) AS avg_imdb_rating,
    COUNT(*) AS count_movie 
  FROM
    imdb_top_1000
  WHERE 
    released_year >= 2010 
  GROUP BY
    released_year 
  HAVING
    COUNT(*) > 8 
  ORDER BY
    avg_imdb_rating DESC LIMIT 5
)
,
trend_meta_score AS 
(
  -- Best movies from critics score
  SELECT
    released_year,
    round(AVG(meta_score), 2) AS avg_meta_score,
    COUNT(*) AS count_movie 
  FROM
    imdb_top_1000
  WHERE 
    released_year >= 2010  
  GROUP BY
    released_year 
  HAVING
    COUNT(*) > 8 
  ORDER BY
    avg_meta_score DESC LIMIT 5
)


SELECT
  genre,
  COUNT(*) AS genre_count 
FROM
  (
    SELECT
      released_year,
      SUBSTRING_INDEX(SUBSTRING_INDEX(genre, ', ', n.digit + 1), ', ', - 1) AS genre -- Extract each genre from the comma-separated list
    FROM
      imdb_top_1000 
      JOIN
        (
          SELECT 0 AS digit 
          UNION ALL
          SELECT 1 
          UNION ALL
          SELECT 2 
          UNION ALL
          SELECT 3
        )
        AS n -- Generates a sequence of numbers the maximum number of genres
        ON LENGTH(genre) - LENGTH(REPLACE(genre, ',', '')) >= n.digit 
  )
  AS genres_table 
WHERE
  released_year IN 
  (
    -- Select the released year of the best movies by user and critic score
    SELECT
      ti.released_year 
    FROM
      trend_imdb_rating AS ti 
      INNER JOIN
        trend_meta_score AS tm 
        ON ti.released_year = tm.released_year 
  )
GROUP BY genre 
ORDER BY genre_count DESC
```

```python
%%sql

WITH trend_imdb_rating AS 
(
  -- Best movies from user rating
  SELECT
    released_year,
    round(AVG(imdb_rating), 2) AS avg_imdb_rating,
    COUNT(*) AS count_movie 
  FROM
    imdb_top_1000
  WHERE 
    released_year >= 2010  
  GROUP BY
    released_year 
  HAVING
    COUNT(*) > 8 
  ORDER BY
    avg_imdb_rating DESC LIMIT 5
)
,
trend_meta_score AS 
(
  -- Best movies from critics score
  SELECT
    released_year,
    round(AVG(meta_score), 2) AS avg_meta_score,
    COUNT(*) AS count_movie 
  FROM
    imdb_top_1000
  WHERE 
    released_year >= 2010  
  GROUP BY
    released_year 
  HAVING
    COUNT(*) > 8 
  ORDER BY
    avg_meta_score DESC LIMIT 5
)
SELECT
  genre,
  SUM(gross) AS revenue 
FROM
  (
    SELECT
      released_year,
      gross,
      -- Get one genre only
      SUBSTRING_INDEX(SUBSTRING_INDEX(genre, ', ', n.digit + 1), ', ', - 1) AS genre -- Extract each genre from the comma-separated list
    FROM
      imdb_top_1000 
      JOIN
        (
          SELECT 0 AS digit 
          UNION ALL
          SELECT 1 
          UNION ALL
          SELECT 2 
          UNION ALL
          SELECT 3 
          UNION ALL
          SELECT 4
        )
        AS n -- Generates sequence of numbers to the maximum number of genres of the movie
        ON LENGTH(genre) - LENGTH(REPLACE(genre, ',', '')) >= n.digit 
  )
  AS genres_table 
WHERE
  released_year IN 
  (
    -- Select the released year of the best movies by user and critic score
    SELECT
      ti.released_year 
    FROM
      trend_imdb_rating AS ti 
      INNER JOIN
        trend_meta_score AS tm 
        ON ti.released_year = tm.released_year 
  )
GROUP BY
  genre 
ORDER BY
  revenue DESC
```
